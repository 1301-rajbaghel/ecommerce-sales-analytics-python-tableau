# Global Sales Analysis Using Python

## Objective
Analyze global sales data to identify revenue trends, profitability drivers, customer behavior, and market performance.


## Import  Libraries

```python
import pandas as pd
```

## Load Dataset

```python
df=pd.read_csv("/users/rajbaghel/desktop/project 2 dataset.csv")
```

## Dataset Overview

```python
df.info()
```

## Data Cleaning

```python
df["sales"] = df["sales"].str.replace(",", "")
df["sales"] = pd.to_numeric(df["sales"])
```

## Dataset Shape

```python
df.shape
```

## Category Analysis

```python
df.groupby("category")["sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("category")["profit"].sum().sort_values(ascending=False)
```

## State Analysis

```python
df.groupby("state")["sales"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("state")["profit"].sum().sort_values(ascending=False).head(10)
```

## Market Analysis

```python
df.groupby("market")["sales"].sum().sort_values(ascending=False)
```

## Year Analysis

```python
df.groupby("year")["sales"].sum()
```

```python
df.groupby("year")["profit"].sum()
```

## Product Analysis

```python
df.groupby("sub_category")["sales"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("sub_category")["profit"].sum().sort_values(ascending=False).head(10)
```

## Customer Analysis

```python
df.groupby("customer_name")["sales"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("customer_name")["profit"].sum().sort_values(ascending=False).head(10)
```

## Shipping Analysis

```python
df.groupby("ship_mode")["sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("ship_mode")["profit"].sum().sort_values(ascending=False)
```

## Correlation Analysis

```python
df[["sales","profit","discount","shipping_cost"]].corr()
```

## KPI Summary

```python
print("Total Sales:", df["sales"].sum())
print("Total Profit:", df["profit"].sum())
print("Total Orders:", df["order_id"].nunique())
```

# Key Findings

1. Technology generated the highest sales and profit.
2. England was the top-performing state.
3. APAC was the highest revenue market.
4. Sales increased consistently from 2011 to 2014.
5. Phones were the top-selling sub-category.
6. Copiers were the most profitable sub-category.
7. Standard Class was the most commonly used shipping mode.
8. Higher discounts negatively impacted profitability.
